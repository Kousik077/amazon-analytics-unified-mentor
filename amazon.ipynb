```python
import pandas as pd
import matplotlib.pyplot as plt

```

```python
data=pd.read_csv("Amazon_Sales_Data.csv ")
```

```python
data
```

```python
data['Order Date']=pd.to_datetime(data['Order Date'])
data['month']=data['Order Date'].dt.month
```

```python
data['year']=data['Order Date'].dt.year
```

```python
print("this graph shows the relationship between units sold on a monthly wise count")
monthly_units_sold=data.groupby('month')['Units Sold'].sum()
plt.figure(figsize=(10,6))
monthly_units_sold.plot(kind='bar')
plt.xlabel('month')
plt.ylabel('Units Sold')
```

```python
print("this graph shows the relationship between units sold on a year wise count")
monthly_units_sold=data.groupby('year')['Units Sold'].sum()
plt.figure(figsize=(10,6))
monthly_units_sold.plot(kind='bar')
plt.xlabel('year')
plt.ylabel('Units Sold')
```
